```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
iris.head()
```

```python
iris_num = iris.select_dtypes(include='number')
iris_normalised_df=(iris_num-iris_num.min())/(iris_num.max()-iris_num.min())
```

```python
flower_name = pd.get_dummies(iris.species)
```

```python
iris_final = pd.concat([iris_normalised_df, flower_name], axis='columns')
```

```python
iris_final.to_csv("./data/iris.csv", index =False)
```
